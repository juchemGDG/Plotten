```python
from pandas import read_csv
from seaborn import set_theme, countplot,scatterplot,boxplot
from matplotlib.pyplot import title, xlabel, ylabel, show,plot,figure
from numpy import polyfit
```

```python
# Daten importieren
df = read_csv("Beispieldaten.csv",sep=";")
df
```

```python
from seaborn import set_theme, countplot
from matplotlib.pyplot import title, xlabel, ylabel, show

set_theme(style="whitegrid")
countplot(data=df,x="Feature2", orient="v")

title("Balkendiagramm f√ºr Feature 2")
xlabel("Feature 2")
ylabel("Anzahl")

show()
```

```python
from seaborn import scatterplot
from matplotlib.pyplot import plot
from numpy import polyfit

# Scatter-Plot erstellen
scatterplot(data=df,x="Feature1",y="Feature2", color="blue", alpha=0.5)
# Trendlinie berechnen (lineare Regression)
x=df["Feature1"]
y=df["Feature2"]
m, b = polyfit(x, y, 1)  # 1 = Gerade (linear)
# Trendlinie plotten
plot(x, m * x + b, color="black")
# Achsen beschriften
xlabel("Wert x")
ylabel("Wert y")
title("Streudiagramm")
# Plot anzeigen
show()
```

```python
from seaborn import boxplot
from matplotlib.pyplot import figure

# Neue Spalte "Gruppe" erstellen
#df["Gruppe"] = where(df["spalte1"] > 6, "viel #Schlaf", "wenig Schlaf")
#Boxplot erstellen
figure(figsize=(8, 6))
#mit Gruppen-Unterteilung
boxplot(x="Target", y="Feature1", data=df, palette=["lightgreen", "green"])
#boxplot(y="Feature1", data=df)
# Achsenbeschriftungen und Titel
title("Boxplot")
xlabel("Gruppe")
ylabel("Feature1")
show()
```

```python
from seaborn import JointGrid,kdeplot,histplot

g = JointGrid(data=df, x="Feature1", y="Feature2", space=0)
g.plot_joint(kdeplot,
             fill=True, clip=((0, 680), (0, 20)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(histplot, color="#03051A", alpha=1, bins=20)
```
